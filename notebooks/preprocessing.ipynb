```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
data_path = "../data/raw/cs-training.csv"
df = pd.read_csv(data_path,  index_col=0)

# Quick overview
df.head()
```

```python
monthly_income_median = df['MonthlyIncome'].median()
df['MonthlyIncome'] = df['MonthlyIncome'].fillna(monthly_income_median)
df['NumberOfDependents'] = df['NumberOfDependents'].fillna(0)
```

```python
# Check for missing values
df.isnull().sum()
```

```python
df['DebtPerDependent'] = df['DebtRatio'] / (df['NumberOfDependents'] + 1)
df['HighUtilization'] = (df['RevolvingUtilizationOfUnsecuredLines'] > 0.8).astype(int)
```

```python
df.to_csv("../data/processed/credit_data_train_processed.csv", index=False)
```
